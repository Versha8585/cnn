```python
from keras.datasets import mnist
import matplotlib.pyplot as plt

# Load dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print("Training data shape:", x_train.shape)
print("Testing data shape:", x_test.shape)

# Visualize first 9 images
plt.figure(figsize=(5,5))
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f"Digit: {y_train[i]}")
    plt.axis('off')
plt.suptitle("MNIST Dataset Sample Images")
plt.show()
```

```python
# ---------- 0. Setup ----------
!pip install -q tensorflow scikit-learn matplotlib

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

tf.random.set_seed(42)
np.random.seed(42)



# Load data
(X_train_full, y_train_full), (X_test, y_test) = mnist.load_data()

print("Training Images Shape:", X_train_full.shape)
print("Training Labels Shape:", y_train_full.shape)
print("Test Images Shape:", X_test.shape)
print("Test Labels Shape:", y_test.shape)

# Quick explanation summary
print("MNIST shapes:", X_train_full.shape, y_train_full.shape, X_test.shape, y_test.shape)
# Example visuals
plt.figure(figsize=(8,4))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_train_full[i], cmap='gray')
    plt.title(f"Label: {y_train_full[i]}")
    plt.axis('off')
plt.suptitle("Sample MNIST Images")
plt.show()

# Class distribution
(unique, counts) = np.unique(y_train_full, return_counts=True)
print("Class distribution (train):", dict(zip(unique, counts)))

# ---------- 2. Preprocess ----------
# Normalize and reshape for CNN
X = X_train_full.astype('float32') / 255.0
X = np.expand_dims(X, -1)   # shape (60000, 28, 28, 1)
X_test_proc = X_test.astype('float32') / 255.0
X_test_proc = np.expand_dims(X_test_proc, -1)

# Split train -> train/val for stacking
X_train, X_val, y_train, y_val = train_test_split(X, y_train_full, test_size=0.15, random_state=42, stratify=y_train_full)
print("Train/Val/Test shapes:", X_train.shape, X_val.shape, X_test_proc.shape)

# ---------- 3. EDA: Correlation matrix via PCA ----------
# Flatten images and compute PCA to reduce dimension for interpretable correlation
X_flat = X_train.reshape(X_train.shape[0], -1)  # (N, 784)
scaler = StandardScaler(with_mean=True, with_std=True)
X_flat_scaled = scaler.fit_transform(X_flat)

# PCA to 30 components
pca = PCA(n_components=30, random_state=42)
X_pca = pca.fit_transform(X_flat_scaled)
print("Explained variance by top 30 PCA components:", np.sum(pca.explained_variance_ratio_))

# Correlation matrix on PCA components (30x30)
corr = np.corrcoef(X_pca.T)
plt.figure(figsize=(8,6))
plt.imshow(corr, cmap='bwr', vmin=-1, vmax=1)
plt.colorbar()
plt.title("Correlation matrix of top-30 PCA components (train set)")
plt.show()

# ---------- Helper: build plotting func ----------
def plot_history(h, title="Model metrics"):
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(h.history['loss'], label='train_loss')
    plt.plot(h.history.get('val_loss', []), label='val_loss')
    plt.legend(); plt.title('Loss')
    plt.subplot(1,2,2)
    plt.plot(h.history['accuracy'], label='train_acc')
    plt.plot(h.history.get('val_accuracy', []), label='val_acc')
    plt.legend(); plt.title('Accuracy')
    plt.suptitle(title)
    plt.show()

# ---------- 4. Build three CNN models (A, B, C) ----------
input_shape = (28,28,1)
num_classes = 10

def build_cnn_a():
    model = models.Sequential([
        layers.Conv2D(32, (3,3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2,2)),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.4),
        layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

def build_cnn_b():
    model = models.Sequential([
        layers.Conv2D(32, (3,3), activation='relu', input_shape=input_shape),
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Conv2D(128, (3,3), activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-3), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

def build_cnn_c():
    model = models.Sequential([
        layers.Conv2D(64, (3,3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2,2)),
        layers.Conv2D(128, (3,3), activation='relu'),
        layers.Flatten(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(5e-4), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# ---------- 5. Train base models (use modest epochs for example) ----------
epochs = 8
batch = 128

cnn_a = build_cnn_a()
cnn_b = build_cnn_b()
cnn_c = build_cnn_c()

print("Training CNN-A")
h_a = cnn_a.fit(X_train, y_train, epochs=epochs, batch_size=batch, validation_data=(X_val, y_val), verbose=2)
plot_history(h_a, "CNN-A training")

print("Training CNN-B")
h_b = cnn_b.fit(X_train, y_train, epochs=epochs, batch_size=batch, validation_data=(X_val, y_val), verbose=2)
plot_history(h_b, "CNN-B training")

print("Training CNN-C")
h_c = cnn_c.fit(X_train, y_train, epochs=epochs, batch_size=batch, validation_data=(X_val, y_val), verbose=2)
plot_history(h_c, "CNN-C training")

# ---------- 6. Evaluate base models on test set ----------
print("Evaluate on test set:")
for name, model in [('CNN-A', cnn_a), ('CNN-B', cnn_b), ('CNN-C', cnn_c)]:
    loss, acc = model.evaluate(X_test_proc, y_test, verbose=0)
    print(f"{name} test acc: {acc:.4f}, loss: {loss:.4f}")

# ---------- 7. Create stacking dataset (use validation set predictions) ----------
# Get probability predictions for validation set (stacking train)
val_pred_a = cnn_a.predict(X_val)  # shape (valN,10)
val_pred_b = cnn_b.predict(X_val)
val_pred_c = cnn_c.predict(X_val)

# Stack them horizontally as features for meta-learner
# Option A: Use probabilities for all classes from each model (concat -> 30 features)
stack_X_train = np.hstack([val_pred_a, val_pred_b, val_pred_c])
stack_y_train = y_val

print("Stacking train features shape:", stack_X_train.shape)

# Build stacking features for test set (meta test)
test_pred_a = cnn_a.predict(X_test_proc)
test_pred_b = cnn_b.predict(X_test_proc)
test_pred_c = cnn_c.predict(X_test_proc)
stack_X_test = np.hstack([test_pred_a, test_pred_b, test_pred_c])
stack_y_test = y_test

# ---------- 8. Train a meta-learner (Logistic Regression) ----------
# Flatten probabilities per class -> multi-class logistic regression
meta = LogisticRegression(multi_class='multinomial', max_iter=1000, solver='saga', random_state=42)
meta.fit(stack_X_train, stack_y_train)

# Evaluate stacking
meta_preds = meta.predict(stack_X_test)
print("Stacked model accuracy:", accuracy_score(stack_y_test, meta_preds))
print(classification_report(stack_y_test, meta_preds))

# ---------- 9. Confusion matrix for stacked model ----------
cm = confusion_matrix(stack_y_test, meta_preds)
plt.figure(figsize=(6,6))
plt.imshow(cm, cmap='Blues')
plt.title('Confusion matrix: Stacked model')
plt.colorbar()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ---------- 10. Save models ----------
cnn_a.save('cnn_a.h5')
cnn_b.save('cnn_b.h5')
cnn_c.save('cnn_c.h5')

import joblib
joblib.dump(meta, 'stacking_meta_lr.joblib')
print("Saved base models and stacking meta-learner.")
```
